```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corpus/train.csv')
```

```python
formal_register = df.formal_register
plt_df = formal_register.value_counts()

fig, ax = plt.subplots()

bar_container = ax.bar(plt_df.index, plt_df.values)

ax.bar_label(bar_container, label_type='center')
plt.title('Formal Register')
plt.xlabel('Notas')
plt.ylabel('Quantidade')

plt.xticks(plt_df.index)
plt.show()
```

```python
print(f'Total de notas: {formal_register.count()}')
print(f'Média das notas: {np.mean(formal_register)}')
print(f'Desvio padrão: {np.std(formal_register)}')
```
